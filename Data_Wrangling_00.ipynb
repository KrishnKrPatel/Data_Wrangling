# WebScraping of data from website

1) urllib.request 
2) Beautiful Soup

```python
import urllib.request as res
```

```python
wiki='https://en.wikipedia.org/wiki/List_of_countries_by_the_number_of_billionaires'
```

```python
page=res.urlopen(wiki)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(page)
```

```python
soup.prettify()
```

```python
soup.title.text
```

```python
soup.find('a')
```

```python
soup.find_all('a')
```

```python
right_table=soup.find('table',class_='wikitable')
right_body=right_table.find('tbody')
rows=right_body.find_all('tr')
```

```python
count=0
data=[]
```

```python
for row in rows:
    if count==0:
        cols=row.find_all('th')
        cols=[element.text.strip() for element in cols]
        data.append([elem for elem in cols if elem])
        count=count+1
    else:
        cols=row.find_all('td')
        cols=[element.text.strip() for element in cols]
        data.append([elem for elem in cols if elem])
```

```python
import pandas as pd
```

```python
dataframe=pd.DataFrame(columns=data[0])
```

```python
for i in range(1,len(data)):
    dataframe=dataframe.append(pd.Series(data[i]))
```
